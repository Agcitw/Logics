Задача 1. 
# Отрисовать график функции mu, которая принимает значение 0 на отрезке [a1; b1] и значение 1 на отрезке [a2; b2].

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def zeroes_ones(i, j, m, k):
  param = 10
  x1 = np.linspace(i, j, param)
  x2 = np.linspace(m, k, param)
  y1 = np.zeros(param)
  y2 = np.ones(param)
  plt.xlabel("x")
  plt.ylabel("mu(x)")
  plt.plot(x1, y1)
  plt.plot(x2, y2)

zeroes_ones(-10, 0.2, 0.5, 10)
```

Задача 2. 
# Реализовать три графика на изображениях выше: график с одной переменной mu(x), график с двумя переменными молодой/старый и график с тремя переменными холодно/комфортно/жарко.


```python
a = 2
b = 33
c = 56
d = 88
n = 1000

def foo(x):
  if x >= a and x <= b:
    return (x - a) / (b - a)
  if x > b and x <= c:
    return 1
  if x > c and x <= d:
    return (d - x) / (d - c)
  return 0

x = np.linspace(0, 100, n)
y = [foo(i) for i in x]
plt.xlabel("x")
plt.ylabel("mu(x)")
plt.plot(x, y)
```

```python
def is_young(x):
  if x >= 0 and x <= 25:
    return 1
  if x > 25 and x <= 100:
    return 1 / (1 + ((x - 25) / 5)**2)
  return -1

def is_old(x):
  if x >= 0 and x <= 50:
    return 0
  if x > 50 and x <= 100:
    return 1 / (1 + ((x - 50) / 5)**(-2))
  return -1

x1 = np.linspace(0, 100, 1000)
y1 = [is_young(i) for i in x]
x2 = np.linspace(0, 100, 1000)
y2 = [is_old(i) for i in x]
plt.xlabel("x")
plt.ylabel("mu(x)")
plt.plot(x1, y1)
plt.plot(x2, y2)
```

```python
def is_cold(t):
  return 1 / (1 + ((t - 10) / 7)**12)

def is_warm(t):
  return 1 / (1 + ((t - 20) / 3)**6)

def is_hot(t):
  return 1 / (1 + ((t - 30) / 6)**10) 

x1 = np.linspace(0, 100, 1000)
y1 = [is_cold(i) for i in x1]
x2 = np.linspace(0, 100, 1000)
y2 = [is_warm(i) for i in x2]
x3 = np.linspace(0, 100, 1000)
y3 = [is_hot(i) for i in x3]
plt.xlabel("t")
plt.ylabel("mu(t)")
plt.plot(x1, y1)
plt.plot(x2, y2) 
plt.plot(x3, y3)
```

#Запрограммировать функции конъюнкции и дизъюнкции (в двух вариантах) и отрицания.

```python
def dis_maximin(a, b):
  return max(a, b)

def kon_maximin(a, b):
  return min(a, b)

def dis_kolorometri(a, b):
  return a + b - a * b

def kon_kolorometri(a, b):
  return a * b

def minus(a):
  return 1 - a
```

# Выполнить задание в задачнике Липатовой с. 53-58, свой вариант(!) с помощью реализованных функций в задаче 3 и реализованных графиков в задаче 2.

```python
def f_a(x):
  if x >= 0 and x <= 3:
    return 0
  if x > 3 and x <= 5:
    return (x - 3) / 2
  if x > 5:
    return 1

def f_b(x):
  if x >= 0 and x <= 3:
    return x / 3
  if x > 3 and x <= 5:
    return (-x + 5) / 2
  if x > 5:
    return 0

def f_c(x):
  if x >= 0 and x <= 5:
    return -x / 5 + 1
  if x > 5 :
    return 0

print("Данные функции:")

a = [f_a(i) for i in np.linspace(0, 9, 10)]
b = [f_b(i) for i in np.linspace(0, 9, 10)]
c = [f_c(i) for i in np.linspace(0, 9, 10)]

plt.plot(np.linspace(0, 9, 10), a)
plt.plot(np.linspace(0, 9, 10), b)
plt.plot(np.linspace(0, 9, 10), c)
plt.xlabel("x")
plt.ylabel("mu")
plt.text(0.2, 1, "mu_c")
plt.text(3, 1, "mu_b")
plt.text(6, 0.95, "mu_a")
```

1) Множество D = A V B ^ C, значит последовательность операций: A V B, (A V B) ^ C. 
2) Далее график функции принадлежности mu_AvB.

```python
#Максиминный метод первая операция 
res1 = []

for i in range(10):
  res1.append(dis_maximin(a[i], b[i]))

print("График максиминным методом:")
plt.plot(np.linspace(0, 9, 10), res1)
plt.xlabel("x")
plt.ylabel("mu_AvB")
```

```python
#Колорометрический метод первая операция
res2 = []

for i in range(10):
  res2.append(dis_kolorometri(a[i], b[i]))

print("График колорометрическим методом:")
plt.plot(np.linspace(0, 9, 10), res2)
plt.xlabel("x")
plt.ylabel("mu")
```

2) Далее график функции принадлежности mu_AvB^C или mu_D.

```python
#Максиминный метод вторая операция
res11 = []

for i in range(10):
  res11.append(kon_maximin(res1[i], c[i]))

print("График максиминным методом:")
plt.plot(np.linspace(0, 9, 10), res11)
plt.xlabel("x")
plt.ylabel("mu_AvB^C or mu_D")
plt.scatter(2, 0.6, color = (1, 0, 0)) #отмечено после вычисления в п.3
```

```python
#Колорометрический метод вторая операция
res22 = []

for i in range(10):
  res22.append(kon_kolorometri(res2[i], c[i]))
  
print("График колорометрическим методом:")
plt.plot(np.linspace(0, 9, 10), res22)
plt.xlabel("x")
plt.ylabel("mu_AvB^C or mu_D")
plt.scatter(2, 0.4, color = (1, 0, 0)) #отмечено после вычисления в п.3
```

3) Ядро множества D состоит из элементов из интервала (0, 5). Выберем элемент 2.

```python
print("mu_a(2) = " + str(a[2]))
print("mu_b(2) = " + str(b[2]))
print("mu_c(2) = " + str(c[2]))
print("Максиминный:")
print("mu_AvB(2) = " + str(dis_maximin(a[2], b[2])))
print("mu_AvB^C(2) = " + str(kon_maximin(dis_maximin(a[2], b[2]), c[2])))
print("Колорометрический:")
print("mu_AvB(2) = " + str(dis_kolorometri(a[2], b[2])))
print("mu_AvB^C(2) = " + str(kon_kolorometri(dis_maximin(a[2], b[2]), c[2])))
```

4) На графиках отмечены точки красным цветом mu_D(2)_maximin и mu_D(2)_kolorometri.

```python
print(res11 == res22)
print(max(res11))
print(max(res22))
```
