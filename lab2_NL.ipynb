Объем закупки

```python
axis3 = [0, 10000, 20000, 30000, 40000, 50000]
volume_low = [1, 1, 0, 0, 0, 0]
volume_normal = [0, 0, 1, 1, 0, 0]
volume_high = [0, 0, 0, 0, 1, 1]
plt.plot(axis3, volume_low, 'r')
plt.plot(axis3, volume_normal, 'g')
plt.plot(axis3, volume_high, 'b')
plt.ylabel("mu")
plt.xlabel("Кол-во")
```

3) Пусть имеется кирпич по 7,5 рублей за штуку качества "3". Нужно определить ориентировочный объем закупки.
Степени уверенности простейших утверждений:
Цена="высокая" - 0
Цена="средняя" - 0,5
Цена="низкая" - 0,5
Качество="высокое" - 0
Качество="среднее" - 1
Качество="низкое" - 0
Степени уверенности посылок правил:
- П1:min(0,0) = 0
- П2:min(0,max(1,0)) = 0
- П3:min(0.5,0) = 0
- П4:min(0.5,1) = 0.5
- П5:min(0.5,0) = 0
- П6:min(0.5,max(0,1)) = 0.5
- П7:min(0.5,0) = 0

Построим новую выходную нечеткую переменную

```python
volume_normal_new = []
for i in volume_normal:
  if i <= 0.5:
    volume_normal_new.append(i)
  else: volume_normal_new.append(0.5)
plt.plot(axis3, volume_normal_new, 'g')

volume_high_new = []
for i in volume_high:
  if i <= 0.5:
    volume_high_new.append(i)
  else: volume_high_new.append(0.5)
plt.plot(axis3, volume_high_new, 'r')

plt.ylabel("mu")
plt.xlabel("Кол-во")
```
