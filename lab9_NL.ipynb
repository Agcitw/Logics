Задание 1. Доказать две формулы: одну из первых четырёх, вторую из вторых четырёх (слайд 47)


Пусть $\mathcal{I} \vDash G(\varphi \cup \psi)$, тогда $\varphi$ всегда становится верным: $\mathcal{I} , i \vDash G\varphi$ для любого i, абсолютно аналогично для $\psi $. Таким образом для каждого i: $\mathcal{I},i \vDash G\varphi \cup G\psi$, как бы всегда выполняется либо $\varphi $ либо $ \psi$, что аналогичо первоначальной записи.

Пусть $\mathcal{I} \vDash \varphi U(\psi \cup \chi)$, тогда существует состояние $\varphi$ момента i, что $\mathcal{I},i \vDash \varphi U \psi$, или состояние $\chi$ момента i, где $\mathcal{I},i \vDash \varphi U \chi$, вследствие того, что в конечный момент оператора U выполняться может как $\psi$, так и $\chi$, а состояние $\varphi$ имеет место или для первого, или для второго состояния соответственно для каждого i до этого, а если во всех моментах до выполняется $\varphi$, то: $\mathcal{I},i \vDash \varphi U \psi \cup \varphi U \chi$, значит формула верна.

### Элементарные события интеллектуальной системы

1. $land_i$ - i-й ЛА садится на полосу \\
2. $takeoff_i$ - i-й ЛА взлетает с полосы \\
3. $busy_i$ - полоса занята i-ым ЛА \\


```python
class Airport:
  def __init__(self, busy : bool):
    self.is_busy = busy


class Plane:
  def __init__(self, on_land : bool, airport : Airport):
    self.is_on_land = on_land
    self.airport = airport

  def run():
    while(true):
      if(not airport.is_busy): #L1
        if(is_on_land):
          airport.is_busy = True #L2
          #взлетаем #L3
          is_on_land = False 
          airport.is_busy = False #L4
        else:
          airport.is_busy = True #L5
          #садимся #L6
          is_on_land = True
          airport.is_busy = False #L7
      else:
        break #ждем полосу

```

```python
SVO = Airport(True)
p1 = Plane(True, SVO)
p1.run()
```
