**Аксиомы дискрипционной логики**

$ЛА - летательный аппарат;
ТС - транспортное средство\\
(1) ЛА = Самолет \cup Вертолет\\
(2) Самолет \cap Вертолет = \emptyset\\
(3) ЛА \subseteq ТС\\
(4) ТС = Пассажирский \cup Грузовой\\
(5) Пассажирский \cap Грузовой = \emptyset\\
(6) Реактивный \subseteq Самолет\\
(7) Винтовой \subseteq Вертолет\\
(8) \neg Планирующий = Реактивный \cup Винтовой\\
$

**Утверждения**

$
  (a) Самолет \cup Вертолет = ТС\\
  (b) Реактивный \cup Винтовой = ЛА\\
  (c) Реактивный \subseteq ЛА\\
  (d) Винтовой \subseteq ЛА\\
  (e) ЛА \subseteq Пассажирский \cup Грузовой\\
  (f) \negПланирующий \subseteq ЛА\\
  (g) \negПассажирский \cap \negГрузовой = \negТС\\
$


**Проверка истинности утверждений**

$
a)\space Из \space(1) \spaceи \space(3): Самолет \cup Вертолет \subseteq ТА.\space Неверно, \spaceт. к. \space TC \ne ЛА\\
b)\space Из \space(6)\space и \space(7): Реактивный \ne Самолет, \spaceВинтовой \ne Вертолет\space (в \space подмножестве \space  могут \space  быть \space  иные \space  элементы) => нельзя \space применить\space (1). \spaceНеверно.\\
c)\space Реактивный \subseteq Самолет \subseteq ЛА. \spaceВерно.\\
d)\space Винтовой \subseteq Вертолет \subseteq ЛА.\space Верно.\\
e)\space ЛА \subseteq ТС = Пассажирский \cup Грузовой. \spaceВерно.\\
f)\space \negПланирующий \subseteq Самолет \cup Вертолет = ЛА. \spaceВерно.\\
g)\space Применяя \space закон \space де \space Моргана \space к \space (8) \space получим \space (g). \space Верно.\\
$
